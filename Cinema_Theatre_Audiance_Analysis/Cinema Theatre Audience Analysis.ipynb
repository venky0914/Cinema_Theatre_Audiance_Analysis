```python
#STEP 1 ====> Fixing data tupes
df.dtypes
```

```python
#changing object data type to =====> categorical
data_types=['Theatre_ID','Show_Time','Movie_Genre','Movie_Type','Snack_Purchased','Customer_Type','Weather','Day_Type']
df[data_types]=df[data_types].astype('category')
#changing data datatype(object) to =====> Datetime
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
```

```python
df.dtypes
```

```python
#STEP 2 ====> REMOVING DUPLICATES
```

```python
#checking duplicates
df.duplicated().sum()
```

```python
#removing duplicates 
df.drop_duplicates(inplace=True)
```

```python
#checking wheather duplicates removed or not
df.duplicated().sum()
```

```python
##STEP 3 ====> FINDING INVALID VALUES AND FILLING THOSE VALUES WITH NAN
```

```python
#Finding invalid values
df.select_dtypes(include='number')\
.where(df.select_dtypes(include='number')<=0)\
.stack()
#invalid values ===> [0,-50,-100,-3]
```

```python
#Filling invalid values with NaN
invalid_num=df.select_dtypes(include='number').columns
df[invalid_num]=df[invalid_num].where(df.select_dtypes(include='number')>0,np.nan)
```

```python
#checking wheather invalid values filled with NaN or Not 
df.select_dtypes(include='number')\
.where(df.select_dtypes(include='number')<=0)\
.stack() 
# values filled with NaN
```

```python
#STEP 4 ====> FINDING SKEW AND OUTLIER 
```

```python
# Finding skew
df.select_dtypes(include='number').skew()
# skew = 0 ====> symmetric
# skew > 0.5 ====> High 
# skew < 0.5 =====> Low
```

```python
# Checking outliers
df.select_dtypes(include='number').boxplot(figsize=(12,5))
```

```python
#STEP 5 ====> FILLING NULL VALUES (NaN) 
```

```python
# I am filling null values with mean because data does't contain any outlier and also skew is low

# Filling numerical columns NaN Values

num_cols=df.select_dtypes(include='number').columns
df[num_cols]=df[num_cols].fillna(df[num_cols].mean())  #=====> Filling with mean

# Filling categorical and object columns NaN Values

cat_cols=df.select_dtypes(include=['object','category']).columns
df[cat_cols]=df[cat_cols].fillna(df[cat_cols].mode().iloc[0]) #======> Filling with mode
```

```python
# creating column 
df['Total_Seats']=df['Seats_Available']+df['Seats_Occupied']
df
```

```python
# Creating group column 
df['Ticket_Price_Range']=pd.cut(df['Ticket_Price_INR'],bins=3,labels=['Low','Medium','High'])
df
```

```python
#creating quarter column
df['Quarter'] = df['Date'].dt.to_period('Q')
df
```

```python
# pie chart of seats occupied by Genre
seats_occ=df.groupby('Movie_Genre')['Seats_Occupied'].sum()
e=[0.2,0,0,0,0]
plt.pie(seats_occ,
        labels=seats_occ.index,
       autopct='%0.f%%',
       shadow=True,
       explode=e
       )
plt.title('Seats Occupied By Genre')
plt.show()
```

```python
# bar graph of seats occupied bt day type
bar_occ=df.groupby('Show_Time')['Seats_Occupied'].sum()
plt.bar(bar_occ.index,bar_occ.values)
plt.xlabel('Show time')
plt.title('Seats occupied by show time')
plt.show()
```

```python
quarter_occ=df.groupby('Quarter')[['Seats_Occupied','Total_Seats','Seats_Available']].sum()
quarter_occ.plot(kind='barh')
plt.title('Quorter wise seats occupation')
plt.show()
```

```python
Customer_Ty=df.groupby('Customer_Type')['Seats_Occupied'].sum()
Customer_Ty.plot(kind='barh')
plt.xlabel('Total_Seats')
plt.title('Total seats occupied by customer type')
plt.show()
```

```python
w=df.groupby('Weather')['Seats_Occupied'].sum()
w.plot(kind='bar')
plt.title('Seats occupied by weather')
plt.show()
```

```python
waiting=df.groupby('Show_Time')['Queue_Time_Minutes'].mean()
waiting.plot(kind='li
ne',color='red',marker='o')
plt.title('Avg Waiting time at snack counter ( Minutes )')
plt.show()
```

```python
snack=df.groupby('Ticket_Price_Range')[['Seats_Occupied','Seats_Available','Total_Seats']].sum()
snack.plot(kind='bar')
plt.title('Ticket occupation based on ticket prize range')
plt.show()
```
